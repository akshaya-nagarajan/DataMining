```python
import pydotplus
from sklearn import tree
from IPython.display import Image, display

clf = tree.DecisionTreeClassifier() 
X = [[100, 1, 1],
    [200, 1, 2],
    [200, 2, 1],
    [200, 2, 2]] 
Y = ['No', 'Yes', 'Yes', 'No']

clf = clf.fit(X, Y)
dot_data = tree.export_graphviz(clf, feature_names=['Windy', 'Air quality', 'Hot'], class_names=['No', 'Yes'], filled=True, rounded=True)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  

# Show graph
Image(graph.create_png())
```
