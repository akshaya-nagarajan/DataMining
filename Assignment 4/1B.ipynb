```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from scipy.stats import expon

wineData = "/Users/akshayanagarajan/Downloads/Assignment/wine.data"
#dataset = pd.read_csv(wineData)
dataset = pd.read_csv(wineData, names=['class','Alcohol','Malic acid','Ash','Alcalinity','Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'diluted wines', 'Proline'])

def equifreq(arr1, m): 
      
    a = len(arr1) 
    n = int(a / m) 
    for i in range(0, m): 
        arr = [] 
        for j in range(i * n, (i + 1) * n): 
            if j >= a: 
                break
            arr = arr + [arr1[j]] 
        print(arr) 
  

def equiwidth(arr1, m): 
    a = len(arr1) 
    w = int((max(arr1) - min(arr1)) / m) 
    min1 = min(arr1) 
    arr = [] 
    for i in range(0, m + 1): 
        arr = arr + [min1 + w * i] 
    arri=[] 
      
    for i in range(0, m): 
        temp = [] 
        for j in arr1: 
            if j > arr[i] and j < arr[i+1]: 
                temp += [j] 
        arri += [temp] 
    print(arri)  
  

i = 1
while i < 13:
    a = dataset.iloc[:,i].values
    
    print("equal frequency binning: Column", i) 
    equifreq(a, 3) 

    print("\nequal width binning: Column", i)       
    equiwidth(a,2)
    i = i + 1



x = dataset.iloc[:,1:2].values
y = dataset.iloc[:,0].values
regression_model = LinearRegression()
regression_model.fit(x, y)
y_predicted = regression_model.predict(x)
print("\nLinear regression")
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)

regression_model = LogisticRegression()
regression_model.fit(x, y)
y_predicted = regression_model.predict(x)
print("\nLogistic regression")
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)


sampleRows = dataset.sample(frac =.25) 
print("\nsampling")
sampleRows 
```
